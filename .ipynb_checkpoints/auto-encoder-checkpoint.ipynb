```python
# 单隐藏层自编码器
from keras.layers import Input, Dense
from keras.models import Model
from keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
 
(x_train, _), (x_test, _) = mnist.load_data()
 
x_train = x_train.astype('float32') / 255. # (60000, 28, 28)
x_test = x_test.astype('float32') / 255.  # (10000, 28, 28)
 
# np.prod()计算所有元素的乘积
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:]))) # (60000, 784)
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))    # (10000, 784)
```

```python
encoding_dim = 32
input_img = Input(shape=(784,))
 
encoded = Dense(encoding_dim, activation='relu')(input_img)  # 784->32
decoded = Dense(784, activation='sigmoid')(encoded)          # 32->784
 
autoencoder = Model(inputs=input_img, outputs=decoded)
encoder = Model(inputs=input_img, outputs=encoded)
 
# 将自编码器拆成两个模型，编码器和解码器
# 解码器的部分需要重新构建输入输出，将自编码器的最后一层作为解码器的网络结构
encoded_input = Input(shape=(encoding_dim,))
decoder_layer = autoencoder.layers[-1]
 
decoder = Model(inputs=encoded_input, outputs=decoder_layer(encoded_input))
 
autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
 
autoencoder.fit(x_train, x_train, epochs=50, batch_size=256, 
                shuffle=True, validation_data=(x_test, x_test))
 
encoded_imgs = encoder.predict(x_test)   # 对测试图片，先进行编码，再解码，看生成的内容是否一样
decoded_imgs = decoder.predict(encoded_imgs)
 
n = 10  # how many digits we will display
# figsize:指定figure的宽和高，单位为英寸
# plt.subplot(numRows,numCols,plotNum)
# 图表的整个绘图区域被分成 numRows 行和 numCols 列,然后按照从左到右，从上到下的顺序对每个子区域进行编号，左上的子区域的编号为1
# plotNum 参数指定创建的 Axes 对象所在的区域
plt.figure(figsize=(20, 4))
for i in range(n):
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
 
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(decoded_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()

# 经过50次迭代以后损失值为 val_loss: loss: 0.1040

# 自编码器也是学习生成与X类似的图片，即产生与X分布一致的数据，自编码器的输入为原始数据，输出为重构后的数据，与原始输入作比较，损失函数为交叉熵
```

```python
# 稀疏自编码器、深层自编码器
input_img = Input(shape=(784,))
encoded = Dense(128, activation='relu')(input_img)    # 784->128->64->32，三层神经网络构成的编码器
encoded = Dense(64, activation='relu')(encoded)
decoded_input = Dense(32, activation='relu')(encoded)
 
decoded = Dense(64, activation='relu')(decoded_input)# 32->64->128->784
decoded = Dense(128, activation='relu')(decoded)
decoded = Dense(784, activation='sigmoid')(encoded)
 
autoencoder = Model(inputs=input_img, outputs=decoded)
encoder = Model(inputs=input_img, outputs=decoded_input)
 
autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
 
autoencoder.fit(x_train, x_train, epochs=50, batch_size=256, 
                shuffle=True, validation_data=(x_test, x_test))
 
encoded_imgs = encoder.predict(x_test)
decoded_imgs = autoencoder.predict(x_test)  # 可以直接对测试数据进行编码解码得到解码结果，不需要专门构建一下解码器，将编码后的结果输入解码器中

#  三层自编码器的损失值比单隐藏层自编码器的损失值更低 loss: 0.0905 重建效果更好一点
```

```python
n = 10 
plt.figure(figsize=(20, 4))
for i in range(n):
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
 
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(decoded_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```
